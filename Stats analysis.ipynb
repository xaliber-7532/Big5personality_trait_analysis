```python
def reverse1(val):  
    for i in range(len(val)):
        if(val[i]== 5):
            val[i] = 1
        elif(val[i] == 4):
            val[i] = 2
        elif(val[i] == 1):
            val[i] = 5
        elif(val[i] == 2):
            val[i] = 4
    return val
```

```python
reverse1(df['AGR1'])
reverse1(df['EXT2'])
reverse1(df['EXT5'])
reverse1(df['EXT7'])
reverse1(df['AGR3'])
reverse1(df['AGR6'])
reverse1(df['AGR8'])
reverse1(df['CON2'])
reverse1(df['CON4'])
reverse1(df['CON5'])
reverse1(df['CON9'])
reverse1(df['NEU2'])
reverse1(df['NEU5'])
reverse1(df['NEU7'])
reverse1(df['OPEN7'])
reverse1(df['OPEN9'])
```

```python
#z test between difference of mean:
#H0: There is no differnce between the means of the score (5 different scores) i.e there is no significant difference between those two scores.
#H1: There is a significant difference between those means i.e between those scores.

```

```python
from scipy import stats
from statsmodels.stats.weightstats import ztest as ztest
```

```python
#level of significance=5% i.e. alpha=0.05,output: (z statistic,p value). 
#if p value is less than level of significance, we can conclude that we have sufficient evidence to reject H0.
#if p value is greater than level of significance, we can conclude that we don't have enough evidence to reject H0.
```
